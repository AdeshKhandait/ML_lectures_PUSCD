# Polynominal Linear Regression

```python
import numpy as np
import  pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
# Independant Variables to train model we take input matrix and not vector
x = dataset.iloc[:, 1:2].values
print(x)
```

```python
# Dependant Variables to train model we take input matrix and not vector
y = dataset.iloc[:, 2].values
print(y)
```

```python
# As are Data is very less we will use all the entire Dataset for train
# And also we do not need feature Scale
```

```python
# Fitting the linear regression to Dataset
from statistics import linear_regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y) 
```

```python
# Fitting the Polynominal Linear Regression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly, y)
```

```python
# Visualising the Linear Regression
plt.scatter(x,y,color = 'red')
plt.plot(x,lin_reg.predict(x), color = 'blue')
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salaries')
plt.show()
```

```python
# Visualising the Polynomial Linear Regression
plt.scatter(x,y,color = 'red')
# plt.plot(x,lin_reg2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.plot(x,lin_reg2.predict(x_poly), color = 'blue')
plt.title('Truth or Bluff(Polynomial regression)')
plt.xlabel('Position Level')
plt.ylabel('Salaries')
plt.show()
```

As we above the there are straight line in between curve we can smooth it with increment it 0.1 

```python
# Visualising the Polynomial Linear Regression and smoothing the curve
x_grid = np.arange(min(x), max(x),0.1) # This will return a vector so we need to reshape
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(x,y,color = 'red')
plt.plot(x_grid,lin_reg2.predict(poly_reg.fit_transform(x_grid)), color = 'blue')
plt.title('Truth or Bluff(Polynomial regression)')
plt.xlabel('Position Level')
plt.ylabel('Salaries')
plt.show()
```

```python
# Predicating the value based Linear Regression
print("Linear Regression : {}".format(lin_reg.predict([[6.5]])))
# 
# Predicating the value based Polynominal Linear Regression
print("Linear Regression : {}".format(lin_reg2.predict(poly_reg.fit_transform([[6.5]]))))
```
